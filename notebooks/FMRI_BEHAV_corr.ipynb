```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import matplotlib
matplotlib.rcParams['axes.linewidth'] = 2
matplotlib.rcParams['xtick.major.size'] = 2
matplotlib.rcParams['xtick.minor.size'] = 1
matplotlib.rcParams['xtick.major.width'] = 2
matplotlib.rcParams['xtick.minor.width'] = 2

matplotlib.rcParams['ytick.major.size'] = 2
matplotlib.rcParams['ytick.minor.size'] = 1
matplotlib.rcParams['ytick.major.width'] = 2
matplotlib.rcParams['ytick.minor.width'] = 2
```

```python
data = pd.read_csv("../data/joined.csv")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True, figsize=(8, 4))

rt = "rt_lag2_clr_pe"

for (sub), group in data.groupby(["sub"]):
    
    col = (sub-1) % 3
    row = (sub-1) // 3
    
    ax = axes[row][col]
    
    x = group[rt].values
    s = x.argsort()
    
    y0 = group["pe_clr2_roi0"].values
    y1 = group["pe_clr2_roi1"].values
    y2 = group["pe_clr2_roi2"].values
    
    ax.set_title("Observer %d (N=%d)." % (sub, len(x)), y=.9)
#     ax.set_title("P%d (r=%0.1f, p(%d)=%0.1f)" % (sub, r, len(y), p), y=.9)
    
    xmin, xmax = (-.4, .4)
    ymin, ymax = (-3000, 3000)
    colors = ["#8855aa", "#55aa88", "#aa5588"]
    rois = ["r-FEF", "l-IPS", "r-IPS"]
    for i, y in enumerate([y0, y1, y2]):
        clr = colors[i]
        a, b, r, p, _ = stats.linregress(x, y)
        ax.plot(x, y, "ko", color=clr, alpha=.75, label=rois[i])
        ax.plot([xmin, xmax], np.array([xmin, xmax])* a + b, ("-" if p<.05 else ":"), linewidth=(3 if p<0.05 else 1), color=clr)

    ax.set_xticks([xmin, xmax])    
    ax.set_ylim(ymin, ymax)
    ax.set_yticks([-2000, 2000])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
ax.legend(loc=4)
        
fig.text(0.5, 0.05, 'RT PE', ha='center', va='center')
fig.text(0.05, 0.5, 'BOLD PE', ha='center', va='center', rotation='vertical')
fig.suptitle("Color lag-2 priming RT vs BOLD PEs")
fig.subplots_adjust(top=.9, hspace=.1, wspace=.05)
```

```python
rt = "rt_lag1_clr_pe"

# Color 1
rois = { 
    "l-FEF": "pe_clr1_roi2",
    "r-FEF": "pe_clr1_roi1", 
    "l-IPS": "pe_clr1_roi0",
    "r-IPS": "pe_clr1_roi3", 
}

fig, axes = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=False, figsize=(len(rois)*1.5, 3.5))
for (sub), group in data.groupby(["sub"]):
    
    col = (sub-1) % 3
    row = (sub-1) // 3
    
    ax = axes[row][col]
    x = group[rt].values
        
    ax.set_title("Obs. %d" % sub, y=.8)
    
    rs, ps = [], []
    nsig = []
    sig = []

    for i, (name, roi) in enumerate(rois.items()):
        clr = colors[i]
        y = group[roi]
        
        a, b, r, p, _ = stats.linregress(x, y)
        
        if p < 0.05:
            sig.append(i)
        else:
            nsig.append(i)
            
        rs.append(r)
        ps.append(p)
        
    ax.bar(sig, [rs[i] for i in sig], width=1, linewidth=1, edgecolor="black", color=colors[2], label="p<0.05")
    ax.bar(nsig, [rs[i] for i in nsig], width=1, linewidth=1, edgecolor="black", color=colors[4], label="n.s.")        
    ax.set_xlim(-.75, len(rs)-1+.75)
    ax.set_xticks(range(len(rs)))
    ax.set_xticklabels(rois, rotation=45)
    ax.set_ylim(-.75, 1.25)
    ax.set_yticks([-0, 1])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    if col != 0:
        ax.set_yticks([])
        ax.spines['left'].set_visible(False)
        
fig.text(0.5, -0.05, 'ROI', ha='center', va='center')
fig.text(0.05, 0.5, 'Pearson r', ha='center', va='center', rotation='vertical')
fig.suptitle("Color lag-1 priming RT vs BOLD PEs")
fig.subplots_adjust(top=.85, hspace=.15, wspace=.05)
fig.savefig("clr1_obs_RTvsBOLD_PE.svg", dpi=300)
```
